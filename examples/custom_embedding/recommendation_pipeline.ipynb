```python
import pandas as pd
import numpy as np
import requests
import json
```

```python
tweet_data = pd.read_csv('../../data/testdata.text_only.csv', index_col=0)
```

```python
tweet_data.head()
```

```python
input = {'text' : tweet_data.text[0]}
response = requests.get('http://0.0.0.0:5000/infer', data=input)
```

```python
matrix_embedding = np.zeros((len(tweet_data), 300))
for i in range(len(tweet_data)):
    text = tweet_data['text'][i]
    input = {'text' : text}
    response = requests.get('http://0.0.0.0:5000/infer', data=input)
    vector_embedding = json.loads(response.text)
    matrix_embedding[i] = vector_embedding
```

```python
embedding_dict = {}
for i in range(len(tweet_data)):
    embedding_dict[i] = matrix_embedding[i].tolist()
```

```python
with open('tweet_embedding.json', 'w') as fp:
    json.dump(embedding_dict, fp)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
input_tweet = 'Hello, I am have a great day and I would like to find someone who feels the same.'
input = {'text' : input_tweet}
response = requests.get('http://0.0.0.0:5000/infer', data=input)
vector_embedding = json.loads(response.text)
```

```python
np.dot(embedding_dict[0], vector_embedding)/(np.linalg.norm(vector_embedding)*np.linalg.norm(embedding_dict[0]))
```

```python
cosine_similarity(np.array(embedding_dict[0]).reshape(1,300), np.array(vector_embedding).reshape(1,300))
```

```python
similarity = cosine_similarity(matrix_embedding, np.array(vector_embedding).reshape(1,300))
```

```python
tweet_data['similarity'] = similarity
```

```python
tweet_data.sort_values(['similarity'], ascending=False)
```
